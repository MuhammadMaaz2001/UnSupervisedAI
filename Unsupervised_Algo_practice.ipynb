**K-Means**

```python
weather = ["sunny", "sunny", "overcast", "rainy", "rainy", "rainy", "overcast", "sunny", "sunny", "rainy",
           'sunny', "overcast", "overcast", "rainy"]
temperature = ["hot", "hot", "hot", 'mild', 'cool', 'cool', 'cool', 'mild', 'cool', 'mild', 'mild', 'mild',
               'hot', 'mild']
play = ["no", "no", "yes", "yes", "yes", "no", "yes", "no", "yes", "yes", "yes", "yes", "yes", "no"]
```

```python
from sklearn.cluster import KMeans
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Weather" :weather,
     "Temperature" : temperature,
     "Play": play
    }
)
df
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
weather_encoded = le.fit_transform(weather)
temperature_encoded = le.fit_transform(temperature)
play_encoded = le.fit_transform(play)

features = list(zip(weather_encoded, temperature_encoded ))
```

```python
df1 = pd.DataFrame(
    {
        "Weather" :weather_encoded,
     "Temperature" : temperature_encoded,
     "Play": play_encoded
    }
)
df1
```

```python
km = KMeans(n_clusters=2)
```

```python
features
```

```python
predicted = km.fit(features).predict(features)
predicted
```

```python
y_predicted = km.fit_predict(df1[["Weather", "Temperature"]])
y_predicted
```

```python
print("Clusters Centers")
df["cluster"] = y_predicted
print(km.cluster_centers_)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df = pd.read_excel('/content/online retail dataset.xlsx')
df.head()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['Invoice'], inplace=True)
df['Invoice'] = df['Invoice'].astype('str')
df = df[~df['Invoice'].str.contains('C')]
df
```

```python
basket  = (df[df['Country']=='United Kingdom']
.groupby(['Invoice','Description'])['Quantity']
.sum().unstack().reset_index().fillna(0)
.set_index('Invoice'))
basket
```

```python
       def encode_units(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
basket_sets = basket.applymap(encode_units)
basket_sets.drop('GLASS STAR FROSTED T-LIGHT HOLDER',inplace = True, axis=1)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >= 6) & (rules['confidence'] >= 0.8) ]
```
